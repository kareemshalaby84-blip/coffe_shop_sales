```python
# Calculate IQR to find the outlier threshold
Q1 = df['unit_price'].quantile(0.25)
Q3 = df['unit_price'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

# Separate the 'Normal' items from 'High-Value' items
high_value_items = df[df['unit_price'] > upper_bound]

print(f"Items priced above ${upper_bound:.2f} are statistical outliers.")
print("Top high-value categories:")
print(high_value_items['product_category'].value_counts())
print(" ")
print(df['product_category'].value_counts())
```

Perform exploratory data analysis (EDA).

```python

# Create a Revenue column (Price * Quantity)
df['total_bill'] = df['transaction_qty'] * df['unit_price']

# Extract Hour, Month, and Day Name
#df['hour'] = pd.to_datetime(df['transaction_time'], format='%H:%M:%S').dt.hour
df['month_name'] = df['transaction_date'].dt.month_name()
df['day_name'] = df['transaction_date'].dt.day_name()

df['Order Year'] = df['transaction_date'].dt.year

df['Order Quarter'] = df['transaction_date'].dt.quarter
```

Analysis

what is the higher total bills per Month 

```python
total_bill_per_month = df.groupby("month_name")['total_bill'].sum().sort_values(ascending=True)
total_bill_per_month.plot(kind = 'bar')
plt.xlabel('month_name')
plt.ylabel('total_bill')
plt.title('total bill per month ', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='x')
```

what is the higher total bills per Quarter

```python
total_bill_per_month = df.groupby("Order Quarter")['total_bill'].sum().sort_values(ascending=False)
total_bill_per_month.plot(kind = 'bar')
plt.xlabel('Quarter Number')
plt.ylabel('total_bill')
plt.title('total bill per Quarter ', fontsize=14, fontweight='bold')
```

what is the higher total bills per Days

```python
total_bill_per_month = df.groupby("day_name")['total_bill'].sum().sort_values(ascending=True)
total_bill_per_month.plot(kind = 'bar')
plt.xlabel('day_name')
plt.ylabel('total_bill')
plt.title('total bill per Days ', fontsize=14, fontweight='bold')

plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='x')
```

```python

cat_rev = df.groupby('product_category')['total_bill'].sum().sort_values(ascending=False)
sns.barplot(x=cat_rev.index, y=cat_rev.values)

plt.xticks(rotation=45)
plt.title('Revenue Contribution by Category')
```

```python

cat_price = df.groupby('product_category')['unit_price'].mean().sort_values(ascending=False)
sns.barplot(x=cat_price.index, y=cat_price.values)

plt.xticks(rotation=45)
plt.title('unit price by Category')
```

```python
store_transactions = df.groupby("store_location")['transaction_qty'].count().sort_values(ascending=True)
store_transactions.plot(kind = 'bar')
plt.xlabel('store_location')
plt.ylabel('transaction_qty')
plt.title('number of visitor per store ', fontsize=14, fontweight='bold')

plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='x')
```

```python
store_transactions = df.groupby("store_location")['transaction_qty'].value_counts()


plt.ylabel('transaction_qty')
plt.title('store_transactions ', fontsize=14, fontweight='bold')

plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='x')


store_transactions.plot(kind='pie', autopct='%1.1f')
plt.title('store_transactions')

plt.show()
```

```python
pivot = df.pivot_table(
    index='day_name',
    columns='store_location',
    values='transaction_id',
    aggfunc='count'
)

pivot.plot(kind='bar', figsize=(10,6))

plt.xlabel('Day')
plt.ylabel('Transactions')
plt.title('Best Day for Each Store')

plt.xticks(rotation=45)

plt.show()
"the best day to visit each store"
```

```python
plt.figure(figsize=(10, 8))
correlation_cols = ['transaction_qty', 'unit_price', 'total_bill']
correlation_matrix = df[correlation_cols].corr()

sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            square=True, linewidths=1, cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Matrix - Numerical Variables', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```
